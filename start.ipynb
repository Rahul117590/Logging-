```python
# factorial program 

#initilsing the logging and connect it with the file name
import logging
logging.basicConfig(filename='myfile.log',level=logging.DEBUG ,
                    format='%(asctime)s-%(levelname)s-%(message)s',force=True)


# now factorial program 
def factorial(n):
    if n<0:
        logging.error("you put the input less than 1 ")
        return None # here we can apply log.error
    elif n==0:
        logging.warning("the factorial of 0 is 1")
        return 1    # here we can apply log.warning
    else:
        result=1
        for i in range(1,n+1): # here we apply log.error
            result *=i
            logging.info(f"factorial of {n} is {result}")
        return result  
#n=int(input("Enter the number: "))
n=0
fact=factorial(n)
print(fact)
if fact is not None:
    print(f"factorial of {n} is {fact}")
else:
    print("ERROR IS FOUND ")
```

```python
import logging
logging.basicConfig(filename="myfile.log",level=logging.DEBUG,
                    format='%(asctime)s-%(levelname)s-%(message)s',force=True)
def addition(*args):# args* it means that there is no limitation of entering the number
    logging.info("From there our funciton is start ")
    sum=0
    for i in args:
        logging.info(f"value of  i is {i}")
        sum=sum+i
    return sum
f=open("myfile.log","r")
print(f.read())
print(addition(1,2,3,4,5,6))
        
```

```python
# calculator type function
import logging
logging.basicConfig(filename="myfile.log",level=logging.DEBUG,
                    format='%(asctime)s-%(levelname)s-%(message)s',force=True)
def operation(op,x,y):

    try:

        if op=="+":
            result=x+y
        elif op=="-":
            result=x-y
        elif op=="*":
            result=x*y
        elif op=="/":
            result=x/y
        else:
            raise ValueError("Invalid operation")
        logging.info(f"Opearation {op} of {x} and {y} is result {result}")
        return result
    except Exception as e:
        logging.error(f'error is found during operation {operation}:{e}')

op="*"
x=10
y=4
# print(operation(op,x,y))
result=operation(op,x,y)
if result is not None:
    print(f" the result of  {x} and {y} by applying operator {op} is {result}")
else:
    print(f"error is found check the log ")
```

Logging Handlers

```python
import logging 
logging.basicConfig(filename='app.log',level=logging.DEBUG,
                    format='%(name)s-%(asctime)s-%(levelname)s-%(message)s',force=True)

file_handler=logging.FileHandler('app.log')
file_handler.setLevel(logging.DEBUG)
formatter=logging.Formatter('%(name)s-%(asctime)s-%(levelname)s-%(message)s') # in which sturcture you wanna store this tell
file_handler.setFormatter(formatter)

#rootLogger
logging.getLogger("").addHandler(file_handler)
```
